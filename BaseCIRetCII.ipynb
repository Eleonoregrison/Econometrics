```python
import pandas as pd

# Charger le fichier CSV
df_CIR_CII = pd.read_csv("CSV_CIR_CII.csv", sep=';')
```

Créer un nouveau DataFrame pour mieux organiser la base de données du CIR et du CII et pouvoir obtenir leurs sorties

```python
import pandas as pd

# Création d'un nouveau DataFrame

# Définition des colonnes
coluna_nom = df_CIR_CII["Nom de l'entreprise"]
coluna_siren = df_CIR_CII["Numéro SIREN"]
coluna_catégorie = df_CIR_CII["Base SIRENE : Catégorie de l'établissement siège"]
coluna_dispositif = df_CIR_CII["Dispositif"]
coluna_AnnéeAgrément = df_CIR_CII["Année d'agrément"]
coluna_pays = df_CIR_CII["Pays"]
coluna_localisation = df_CIR_CII["Localisation"]
coluna_activité = df_CIR_CII["Activité"]

df_CI = pd.DataFrame({
    'Nom': coluna_nom,
    'SIREN': coluna_siren,
    'catégorie': coluna_catégorie,
    'Dispositif': coluna_dispositif,
    "Année d'agrément": coluna_AnnéeAgrément,
    'Pays': coluna_pays,
    'Localisation': coluna_localisation,
    'Activité': coluna_activité
})


# Créer un DataFrame uniquement avec les entreprises françaises
df_CIfr = df_CI[df_CI['Pays'] == 'France']


# Calculer le nombre d'agréments avec des entreprises françaises et étrangères
Agre = len(df_CI)
nCIR = df_CI[df_CI['Dispositif'] == 'CIR'].shape[0]
nCII = df_CI[df_CI['Dispositif'] == 'CII'].shape[0]
Agre_Nfr = len(df_CI) - len(df_CIfr)
Agre_fr = len(df_CIfr)
per_fr = (Agre_fr/Agre) * 100

print(f'Agréments: {Agre:,}')
print(f'           {nCIR:,} CIR ({((nCIR/Agre)*100):.2f}%) et {nCII:,} CII ({((nCII/Agre)*100):.2f}%)')
print(f'Agréments avec des entreprises étrangères: {Agre_Nfr:,} ({((Agre_Nfr/Agre)*100):.2f}%)')
print(f'Agréments avec des entreprises françaises: {Agre_fr:,} ({((Agre_fr/Agre)*100):.2f}%)')

df_CIfr
```

Il y a des entreprises françaises avec le même nom, mais des numéros SIREN différents.

```python
# Regrouper par nom de l'entreprise et compter le nombre de SIRENs uniques
entreprises_sirens_differents = df_CIfr.groupby('Nom')['SIREN'].nunique()
# Filtrer uniquement les entreprises avec plus d'un SIREN
entreprises_avec_sirens_differents = entreprises_sirens_differents[entreprises_sirens_differents > 1]
# Compter combien d'entreprises ont le même nom mais des SIRENs différents
nb_entreprises_sirens_differents = entreprises_avec_sirens_differents.shape[0]
# Afficher le résultat
print(f"Il y a {nb_entreprises_sirens_differents} entreprises avec le même nom, mais des SIRENs différents.")

# Regrouper par nom, localisation et activité, et vérifier s'il y a plus d'un numéro SIREN pour chaque combinaison
empresas_com_siren_diferente = df_CIfr.groupby(['Nom', 'Localisation', 'Activité'])['SIREN'].nunique()
# Filtrer les combinaisons qui ont plus d'un numéro SIREN associé
empresas_diferentes = empresas_com_siren_diferente[empresas_com_siren_diferente > 1]
# Compter combien d'entreprises ont le même nom mais des SIRENs différents
nb_entreprises_siren_different = empresas_diferentes.shape[0]
# Afficher le résultat
print(f"Il y a {nb_entreprises_siren_different} entreprises avec le même nom, localisation et activité mais des SIRENs différents.")
```

Regroupement des entreprises avec le même SIREN dans un DataFrame.

```python
# Fonction pour combiner les informations d'une colonne, séparées par une nouvelle ligne et un point-virgule
def combiner_informations_nouvelle_ligne(colonne):
    uniques = colonne.unique()  # Obtenir les valeurs uniques dans la colonne
    if len(uniques) > 1:  # Si plusieurs valeurs uniques existent, on les combine
        return '; '.join(map(str, uniques))  # Séparer par une nouvelle ligne et un point-virgule
    else:
        return str(uniques[0])  # Si toutes les valeurs sont identiques, on garde la première

# Ajouter la colonne 'Nombre d'agréments' pour compter les répétitions
df_agrupado_par_siren = df_CIfr.groupby('SIREN').agg({
    'Nom': combiner_informations_nouvelle_ligne,  # Combiner les noms s'ils sont différents
    'catégorie': combiner_informations_nouvelle_ligne,  # Combiner les catégories s'il y a des différences
    'Dispositif': lambda x: '; '.join(x),  # Combiner tous les dispositifs séparés par ;
    "Année d'agrément": lambda x: '; '.join(map(str, x)),  # Combiner tous les années d'agrément séparés par ;
    'Pays': combiner_informations_nouvelle_ligne,  # Combiner les pays (dans votre cas, ce sera toujours 'France')
    'Localisation': combiner_informations_nouvelle_ligne,  # Combiner les localisations
    'Activité': combiner_informations_nouvelle_ligne  # Combiner les activités
}).reset_index()

# Calculer le nombre d'agréments (répétitions) pour chaque SIREN et l'ajouter au dataframe
df_agrupado_par_siren['Nombre d\'agréments'] = df_CIfr.groupby('SIREN').size().values

# Réorganiser les colonnes pour que 'Nombre d'agréments' soit avant 'Année d'agrément'
colonnes_ordre = ['SIREN', 'Nom', 'catégorie', 'Nombre d\'agréments', 'Dispositif', "Année d'agrément", 'Pays', 'Localisation', 'Activité']
df_agrupado_par_siren = df_agrupado_par_siren[colonnes_ordre]

# Afficher le nouveau dataframe
df_agrupado_par_siren

```

Vérification s'il y a des entreprises avec le même SIREN mais un nom différent :

```python

# Regrouper par SIREN et vérifier si les entreprises ont des noms différents
entreprises_noms_diff = df_CIfr.groupby('SIREN')['Nom'].nunique()

# Filtrer les SIRENs qui ont plus d'un nom associé
siren_avec_noms_diff = entreprises_noms_diff[entreprises_noms_diff > 1].index

# Filtrer les entreprises avec ces SIRENs
entreprises_avec_noms_diff = df_CIfr[df_CIfr['SIREN'].isin(siren_avec_noms_diff)]

# Afficher les entreprises avec le même SIREN mais des noms différents
print(entreprises_avec_noms_diff[['SIREN', 'Nom']].drop_duplicates())

```

Il n'y en a qu'une, alors nous allons changer le nom d'INNOVALTECH en LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET, qui selon la consultation faite dans la base SIRENE est le nom correct de l'entreprise.

```python
# Fonction pour remplacer 'INNOVALTECH' par 'LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET' et retirer les ';'
def corriger_nom(colonne):
    # Remplacer les noms 'INNOVALTECH' par 'LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET'
    return colonne.replace('INNOVALTECH', 'LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET').replace(';', '')

# Appliquer la fonction uniquement aux lignes qui contiennent 'INNOVALTECH' ou 'LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET'
df_agrupado_par_siren['Nom'] = df_agrupado_par_siren['Nom'].apply(lambda x: corriger_nom(x) if 'INNOVALTECH' in x or 'LYCEE GENERAL ET TECHNOLOGIQUE CONDORCET' in x else x)
```
