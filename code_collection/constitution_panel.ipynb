# Base de données brevets

```python
df_patents = pd.read_csv('brevet_total.CSV')
df_patents
```

```python
# on prend en compte les family ID et non les brevets
df_grouped = df_patents.groupby('Famille ID').agg({
    'SIREN': 'first',  # Prendre la première valeur (supposons que le SIREN est identique pour chaque famille)
    'Brevet': lambda x: ', '.join(x),  # Joindre les brevets par une virgule
    'Année de Dépôt': 'min',  # Même logique pour l'année de dépôt

}).reset_index()
df_grouped
```

```python
# Définir les années pour lesquelles on souhaite créer les colonnes
years = list(range(2016, 2025))

# Créer une colonne pour chaque année avec des valeurs par défaut à 0
for year in years:
    df_grouped[year] = 0

# Mettre un 1 dans la colonne de l'année de dépôt correspondante
for index, row in df_grouped.iterrows():
    year = row['Année de Dépôt']
    if year in years:
        df_grouped.at[index, year] = 1
```

```python
df_grouped.head(10)
```

```python
#on compte le nombre de family ID par an par siren 
df_grouped = df_grouped.groupby('SIREN').agg(
    {**{col: 'sum' for col in [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]},  # Somme des colonnes CIR
     'Famille ID': lambda x: ', '.join(map(str, x)),  # Concaténation des Famille ID après conversion en str
     'Année de Dépôt': lambda x: ', '.join(map(str, x))  # Concaténation des Année de Dépôt après conversion en str
    }).reset_index()

df_grouped.rename(columns={'SIREN': 'siren', 2016 : 'brevets2016', 2017 : 'brevets2017', 2018 : 'brevets2018', 2019 : 'brevets2019', 2020 : 'brevets2020', 2021 : 'brevets2021', 2022 : 'brevets2022', 2023 : 'brevets2023', 2024 : 'brevets2024'}, inplace=True)
df_grouped.head()
```

```python
df_grouped['siren'] = df_grouped['siren'].astype(str).str.replace(' ', '').astype(int)
```

```python
df_merged = pd.merge(df_grouped, df_CIR, on='siren', how='right')
df_merged['categorieEntreprise']
```

```python
df_merged['Id_ETI'] = df_CIR['categorieEntreprise'].apply(lambda x: 0 if x == 'GE' else 1)
df_merged= df_merged.drop(columns = ['categorieEntreprise'])
df_merged['dateCreationUniteLegale'] = pd.to_datetime(df_merged['dateCreationUniteLegale']).dt.year
# Liste des colonnes où remplacer les NaN
colonnes_brevets = [f'brevets{annee}' for annee in range(2016, 2025)]

# Remplacer les NaN par 0 dans les colonnes spécifiées
df_merged[colonnes_brevets] = df_merged[colonnes_brevets].fillna(0)
# On ne garde que que les année 2016-2020
df_merged = df_merged.drop(columns = ['brevets2021', 'brevets2022', 'brevets2023', 'brevets2024', 'CIR2021', 'CIR2022', 'CIR2023', 'Famille ID', 'Année de Dépôt', 'activitePrincipaleUniteLegale', 'somme_CIR']) 
df_merged
```

**panel filtré avec les valeurs du CIR**

On ne garde que les entreprises qui reçoivent du CIR au moins une fois sur la periode, qui produisent des brevets et qui reçoivent du CIR sur une période continue

```python
# Regrouper les colonnes brevets et CIR sur la période 2016-2020
brevets_cols = [f"brevets{year}" for year in range(2016, 2021)]
cir_cols = [f"CIR{year}" for year in range(2016, 2021)]

# Étape 1 : Détecter les entreprises avec brevets = 0 sur toute la période
df_merged["brevets_zero"] = df_merged[brevets_cols].apply(lambda row: all(val == 0 for val in row), axis=1)

# Étape 2 : Détecter les entreprises avec CIR = 0 sur toute la période
df_merged["cir_zero"] = df_merged[cir_cols].apply(lambda row: all(val == 0 for val in row), axis=1)

# Étape 3 : Détecter les entreprises avec CIR en dents de scie
def est_dent_de_scie(row):
    indices_non_zero = [i for i, val in enumerate(row) if val != 0]
    if len(indices_non_zero) <= 1:
        return False
    return all(j - i > 1 for i, j in zip(indices_non_zero, indices_non_zero[1:]))

df_merged["cir_dent_de_scie"] = df_merged[cir_cols].apply(est_dent_de_scie, axis=1)

# Étape 4 : Supprimer les lignes problématiques
df_filtre = df_merged[~(df_merged["brevets_zero"] | df_merged["cir_zero"] | df_merged["cir_dent_de_scie"])]

# Étape 5 : Supprimer les colonnes auxiliaires
df_filtre = df_filtre.drop(columns=["brevets_zero", "cir_zero", "cir_dent_de_scie"])
df_filtre
```

Nous avons ensuite corrigé les données à la main pour finalement obtenir le fichier : database238.csv

```python
""" Création du panel à partir de nos data """
import pandas as pd
df = pd.read_csv("database238.csv") # pour avoir ce panel pour la database de 2500 siren, remplacer par finalbase.csv
df

# Liste des colonnes de brevets (2016 à 2020)
brevets_columns = [f'brevets{year}' for year in range(2016, 2021)]

# Liste des colonnes CIR (2016 à 2020)
cir_columns = [f'CIR{year}' for year in range(2016, 2021)]

# Transformation du DataFrame en panel (format long) avec la fonction melt pour les brevets
df_brevets_panel = pd.melt(df, id_vars=['siren', 'division', 'dateCreationUniteLegale'], 
                   value_vars=brevets_columns, 
                   var_name='annee', value_name='brevets')

# Transformation du DataFrame en panel (format long) avec la fonction melt pour les CIR
df_CIR_panel = pd.melt(df, id_vars=['siren', 'division', 'dateCreationUniteLegale'], 
                       value_vars=cir_columns, 
                       var_name='annee', value_name='CIR')

# Nettoyage des colonnes d'année dans les deux panels
df_brevets_panel['annee'] = df_brevets_panel['annee'].apply(lambda x: x.replace('brevets', ''))
df_CIR_panel['annee'] = df_CIR_panel['annee'].apply(lambda x: x.replace('CIR', ''))

# Fusion des deux panels pour avoir à la fois les informations sur les brevets et le CIR
df_panel = pd.merge(df_brevets_panel, df_CIR_panel, on=['siren', 'division', 'dateCreationUniteLegale', 'annee'], how='left')

# Ajout de la colonne 'idEti' basée sur la colonne 'Id_ETI' (avec 1 pour ETI, 0 pour GE)
# Assurez-vous que la colonne 'Id_ETI' contient des valeurs comme 'ETI' ou 'GE'
df['idEti'] = df['categorieEntreprise'].apply(lambda x: 1 if x == 1 else (0 if x == 0 else None))

# Ajout de la colonne 'idEti' au panel
df_panel = pd.merge(df_panel, df[['siren', 'idEti']], on='siren', how='left')




# Sauvegarde du panel dans un fichier CSV
df_panel.to_csv("panel_finalbase.csv", index=False)

# Affichage du résultat
df_panel
```

**Modèle linéaire :**

```python
model = smf.ols('CIR ~ C(division,Treatment("72")) + C(annee) + Id_ETI + annee_creation + RetD', data=panel_0_1)

# Ajuster le modèle
result = model.fit()

# Résumé des résultats
print(result.summary())
```

**Modèle logit :**

Le modèle parvient à converger pour la régression : 
$$ CIR \sim  année +Id_{ETI} + annee_creation+ R\&D$$
Mais il ne parvient pas à converger lorsque l'on rajoute le secteur d'activité dans la regression.

```python
model = smf.logit('CIR ~  C(division,Treatment("72"))+ C(annee) + Id_ETI + annee_creation+ RetD', data=panel_0_1)

# Ajuster le modèle
result = model.fit(maxiter=500, tol=1e-5)

# Résumé des résultats
print(result.summary())
```

Essayons de réduire le nombre de secteurs : 

**Modèle probit :**

```python
model = smf.probit('CIR ~ C(division,Treatment("72")) + C(annee) + Id_ETI + annee_creation + RetD', data=panel_0_1)

# Ajuster le modèle
result = model.fit()

# Résumé des résultats
print(result.summary())
```
