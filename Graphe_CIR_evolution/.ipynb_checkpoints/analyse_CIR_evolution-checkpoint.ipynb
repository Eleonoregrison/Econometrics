```python
#comparaison 
# Calculer les moyennes
df_filtered = df[df['annee'] >= 2015]

avg_ETI_beneficiaire = df_filtered['ETI_beneficiaire_nbr'].mean()
avg_GE_beneficiaire = df_filtered['GE_beneficaire_nbr'].mean()
avg_ETI_dps = df_filtered['ETI_dps_declare'].mean()
avg_GE_dps = df_filtered['GE_dps_declare'].mean()
avg_ETI_CIR = df_filtered['ETI_CIR'].mean()
avg_GE_CIR = df_filtered['GE_CIR'].mean()

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bénéficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.bar(['ETI', 'GE'], [avg_ETI_beneficiaire, avg_GE_beneficiaire], color=['blue', 'orange'])
plt.title("Moyenne des Bénéficiaires d'ETI et GE (2015-2021)")
plt.ylabel("Nombre de Bénéficiaires")
plt.grid(axis='y')

# Graphique 2 : Moyenne des dépenses déclarées d'ETI et GE
plt.subplot(1, 3, 2)
plt.bar(['ETI', 'GE'], [avg_ETI_dps, avg_GE_dps], color=['blue', 'orange'])
plt.title("Moyenne des Dépenses Déclarées d'ETI et GE (2015-2021)")
plt.ylabel("Dépenses Déclarées")
plt.grid(axis='y')

# Graphique 3 : Moyenne des montants de CIR reçus
plt.subplot(1, 3, 3)
plt.bar(['ETI', 'GE'], [avg_ETI_CIR, avg_GE_CIR], color=['blue', 'orange'])
plt.title("Moyenne des Montants de CIR Reçus (2015-2021)")
plt.ylabel("Montant de CIR")
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Calculer le taux de croissance de bénéficiaires
df['ETI_b_growth'] = df['ETI_beneficiaire_nbr'].pct_change() * 100
df['GE_b_growth'] = df['GE_beneficaire_nbr'].pct_change() * 100

#Calculer le taux de croissance des dps déclaré
df['ETI_d_growth'] = df['ETI_dps_declare'].pct_change() * 100
df['GE_d_growth'] = df['GE_dps_declare'].pct_change() * 100

# Calculer le taux de croissance de la créance recherche
df['ETI_CIR_growth'] = df['ETI_CIR'].pct_change() * 100
df['GE_CIR_growth'] = df['GE_CIR'].pct_change() * 100

#pour n'avoir que les années sur lesquelles on a des données 
df_filtered = df[df['annee'] >= 2015]

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bénéficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_b_growth'][1:], marker='o', label='Taux de Croissance des Bénéficiaires ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_b_growth'][1:], marker='o', label='Taux de Croissance des Bénéficiaires GE (%)', color='orange')

# Ajuster l'échelle de l'axe X pour commencer à partir de 2015
plt.xlim(2015, 2021)
# Ajouter des détails au graphique
plt.title("Évolution du Taux de Croissance des bénéficiaires")
plt.xlabel("Année")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les années à partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 2 : Moyenne des dépenses déclarées d'ETI et GE
plt.subplot(1, 3, 2)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_d_growth'][1:], marker='o', label='Taux de Croissance des dépenses ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_d_growth'][1:], marker='o', label='Taux de Croissance des dépenses GE (%)', color='orange')

# Ajuster l'échelle de l'axe X pour commencer à partir de 2015
plt.xlim(2015, 2021)
# Ajouter des détails au graphique
plt.title("Évolution du Taux de Croissance des dépenses déclarés")
plt.xlabel("Année")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les années à partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 3 : Moyenne des montants de CIR reçus
plt.subplot(1, 3, 3)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_CIR_growth'][1:], marker='o', label='Taux de Croissance du CIR des ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_CIR_growth'][1:], marker='o', label='Taux de Croissance du CIR GE (%)', color='orange')

# Ajuster l'échelle de l'axe X pour commencer à partir de 2015
plt.xlim(2015, 2021)
# Ajouter des détails au graphique
plt.title("Évolution du Taux de Croissance du CIR ")
plt.xlabel("Année")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les années à partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()


plt.show()
```

```python
""" On constate des dynamiques distinctes entre GE et ETI entre 2017 et 2019 donc analyse de plus pour comparer année par année """

import matplotlib.pyplot as plt
df_filtered = df[df['annee'] >= 2014]

# Moyennes
avg_ETI_beneficiaire = df_filtered['ETI_beneficiaire_nbr'].mean()
avg_GE_beneficiaire = df_filtered['GE_beneficaire_nbr'].mean()
avg_ETI_dps = df_filtered['ETI_dps_declare'].mean()
avg_GE_dps = df_filtered['GE_dps_declare'].mean()
avg_ETI_CIR = df_filtered['ETI_CIR'].mean()
avg_GE_CIR = df_filtered['GE_CIR'].mean()

# Normalisation des données par la moyenne
df_normalized = df_filtered.copy()
df_normalized['ETI_beneficiaire_nbr'] = df_filtered['ETI_beneficiaire_nbr'] - avg_ETI_beneficiaire
df_normalized['GE_beneficaire_nbr'] = df_filtered['GE_beneficaire_nbr'] - avg_GE_beneficiaire
df_normalized['ETI_dps_declare'] = df_filtered['ETI_dps_declare'] - avg_ETI_dps
df_normalized['GE_dps_declare'] = df_filtered['GE_dps_declare'] - avg_GE_dps
df_normalized['ETI_CIR'] = df_filtered['ETI_CIR'] - avg_ETI_CIR
df_normalized['GE_CIR'] = df_filtered['GE_CIR'] - avg_GE_CIR

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bénéficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.plot(df_filtered['annee'], df_normalized['ETI_beneficiaire_nbr'], marker='o', label='Nombre de bénéficiaires ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_beneficaire_nbr'], marker='o', label='Nombre de bénéficiaires GE', color='orange')
plt.xlim(2016, 2020)
plt.title("Évolution normalisée des bénéficiaires")
plt.xlabel("Année")
plt.ylabel("Nombre de bénéficiaires")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 2 : Moyenne des dépenses déclarées d'ETI et GE
plt.subplot(1, 3, 2)
plt.plot(df_filtered['annee'], df_normalized['ETI_dps_declare'], marker='o', label='Dépenses déclarés des ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_dps_declare'], marker='o', label='Dépenses déclarés des GE', color='orange')
plt.xlim(2016, 2020)
plt.title("Évolution normalisée des dépenses déclarées")
plt.xlabel("Année")
plt.ylabel("Dépenses en M€")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 3 : Moyenne des montants de CIR reçus
plt.subplot(1, 3, 3)
plt.plot(df_filtered['annee'], df_normalized['ETI_CIR'], marker='o', label='CIR des ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_CIR'], marker='o', label='CIR des GE', color='orange')
plt.xlim(2016, 2020)
plt.title("Évolution normalisée du CIR reçu")
plt.xlabel("Année")
plt.ylabel("Créances en M€")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
NAF_dict = {
	"C20_C21": "Pharmacie, parfumerie et entretien",
	"C30": "Construction navale, aéronautique et ferroviaire",
	"C29": "Industrie automobile",
	"C26": "Industrie électrique et électronique",
	"C22": "Chimie, caoutchouc, plastiques",
	"C25": "Industrie mécanique",
	"C24": "Métallurgie et transformation des métaux",
	"C19": "Hydrocarbures, production d'énergie",
	"C13_C14_C15": "Textile, habillement, cuir",
	"C10": "Industrie agricole et alimentaire",
	"C32": "Autres industries manufacturières",
	"C33": "Gestion des holdings industriels",
	"J62": "Conseil et assistance en informatique",
	"M71": "Services d'architecture et d'ingénierie",
	"G": "Commerce",
	"M70": "Conseil et assistance aux entreprises",
	"J61": "Services de télécommunications",
	"K": "Services bancaires et assurances",
	"M72": "Recherche et développement",
	"S": "Autres services", 
	"A_F": "Autres secteurs" #(Agriculture, sylviculture, pêche et Bâtiment, travaux publics)
}
```

```python
# Charger le fichier CSV
df = pd.read_csv("sector_CIR_evolution.csv", sep=';')

# Supprimer les espaces des noms de colonnes
df.columns = df.columns.str.strip()
# Remplacer les NaN par 0
df.fillna(0, inplace=True)
# Remplacer les virgules par des points dans les colonnes, si nécessaire
df = df.replace(',', '.', regex=True)
# Convertir les colonnes en numériques, en forçant les erreurs à NaN
df = df.apply(pd.to_numeric, errors='coerce')

means= {}
for col in df.columns[1:] : 
    means[col] = df[col].mean()

# Initialiser les listes pour les labels et les valeurs
labels = []
values = []

# Extraire les clés et valeurs
for key, value in means.items():
    matched = False  # Indicateur pour vérifier si une correspondance a été trouvée
    for naf_key, naf_value in NAF_dict.items():
        if naf_key in key:
            labels.append(naf_value)  # Ajouter le nom correspondant du dict NAF
            values.append(value)  # Ajouter la valeur correspondante
            matched = True
            break
    if not matched:
        print(f"Aucune correspondance trouvée pour la clé : {key}")  # Pour le débogage

# Vérifier les longueurs des listes
print(f"Nombre de labels : {len(labels)}")
print(f"Nombre de valeurs : {len(values)}")


#Affichage
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='skyblue')

# Ajouter un titre et des labels
plt.title('Moyennes des part de CIR à chaque secteur entre 2009 et 2021', fontsize=14)
plt.ylabel('Moyenne des parts de CIR', fontsize=12)

# Faire pivoter les labels de l'axe x pour une meilleure lisibilité
plt.xticks(rotation=45, ha='right')

# Afficher le diagramme
plt.tight_layout()
plt.show()
```

```python
#Filtrer les colonnes avec des moyennes supérieures à 8%
filtered_columns = [key for key, value in means.items() if value > 8]

# Afficher l'évolution pour les colonnes filtrées
plt.figure(figsize=(12, 8))

for col in filtered_columns:
    plt.plot(df['annee'], df[col], marker='o', label=col)
    
# Ajouter des titres et des labels
plt.title('Évolution des parts de CIR pour les secteurs avec des moyennes > 8%', fontsize=16)
plt.xlabel('Année', fontsize=12)
plt.ylabel('Part de CIR', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()

# Afficher le graphique
plt.show()
```
