```python
#comparaison 
# Calculer les moyennes
df_filtered = df[df['annee'] >= 2015]

avg_ETI_beneficiaire = df_filtered['ETI_beneficiaire_nbr'].mean()
avg_GE_beneficiaire = df_filtered['GE_beneficaire_nbr'].mean()
avg_ETI_dps = df_filtered['ETI_dps_declare'].mean()
avg_GE_dps = df_filtered['GE_dps_declare'].mean()
avg_ETI_CIR = df_filtered['ETI_CIR'].mean()
avg_GE_CIR = df_filtered['GE_CIR'].mean()

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bnficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.bar(['ETI', 'GE'], [avg_ETI_beneficiaire, avg_GE_beneficiaire], color=['blue', 'orange'])
plt.title("Moyenne des Bnficiaires d'ETI et GE (2015-2021)")
plt.ylabel("Nombre de Bnficiaires")
plt.grid(axis='y')

# Graphique 2 : Moyenne des dpenses dclares d'ETI et GE
plt.subplot(1, 3, 2)
plt.bar(['ETI', 'GE'], [avg_ETI_dps, avg_GE_dps], color=['blue', 'orange'])
plt.title("Moyenne des Dpenses Dclares d'ETI et GE (2015-2021)")
plt.ylabel("Dpenses Dclares")
plt.grid(axis='y')

# Graphique 3 : Moyenne des montants de CIR reus
plt.subplot(1, 3, 3)
plt.bar(['ETI', 'GE'], [avg_ETI_CIR, avg_GE_CIR], color=['blue', 'orange'])
plt.title("Moyenne des Montants de CIR Reus (2015-2021)")
plt.ylabel("Montant de CIR")
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Calculer le taux de croissance de bnficiaires
df['ETI_b_growth'] = df['ETI_beneficiaire_nbr'].pct_change() * 100
df['GE_b_growth'] = df['GE_beneficaire_nbr'].pct_change() * 100

#Calculer le taux de croissance des dps dclar
df['ETI_d_growth'] = df['ETI_dps_declare'].pct_change() * 100
df['GE_d_growth'] = df['GE_dps_declare'].pct_change() * 100

# Calculer le taux de croissance de la crance recherche
df['ETI_CIR_growth'] = df['ETI_CIR'].pct_change() * 100
df['GE_CIR_growth'] = df['GE_CIR'].pct_change() * 100

#pour n'avoir que les annes sur lesquelles on a des donnes 
df_filtered = df[df['annee'] >= 2015]

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bnficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_b_growth'][1:], marker='o', label='Taux de Croissance des Bnficiaires ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_b_growth'][1:], marker='o', label='Taux de Croissance des Bnficiaires GE (%)', color='orange')

# Ajuster l'chelle de l'axe X pour commencer  partir de 2015
plt.xlim(2015, 2021)
# Ajouter des dtails au graphique
plt.title("volution du Taux de Croissance des bnficiaires")
plt.xlabel("Anne")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les annes  partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 2 : Moyenne des dpenses dclares d'ETI et GE
plt.subplot(1, 3, 2)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_d_growth'][1:], marker='o', label='Taux de Croissance des dpenses ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_d_growth'][1:], marker='o', label='Taux de Croissance des dpenses GE (%)', color='orange')

# Ajuster l'chelle de l'axe X pour commencer  partir de 2015
plt.xlim(2015, 2021)
# Ajouter des dtails au graphique
plt.title("volution du Taux de Croissance des dpenses dclars")
plt.xlabel("Anne")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les annes  partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 3 : Moyenne des montants de CIR reus
plt.subplot(1, 3, 3)
plt.plot(df_filtered['annee'][1:], df_filtered['ETI_CIR_growth'][1:], marker='o', label='Taux de Croissance du CIR des ETI (%)', color='blue')
plt.plot(df_filtered['annee'][1:], df_filtered['GE_CIR_growth'][1:], marker='o', label='Taux de Croissance du CIR GE (%)', color='orange')

# Ajuster l'chelle de l'axe X pour commencer  partir de 2015
plt.xlim(2015, 2021)
# Ajouter des dtails au graphique
plt.title("volution du Taux de Croissance du CIR ")
plt.xlabel("Anne")
plt.ylabel("Taux de Croissance (%)")
plt.xticks(df_filtered['annee'][1:])  # Afficher les annes  partir de 2015
plt.grid()
plt.legend()
plt.tight_layout()


plt.show()
```

```python
""" On constate des dynamiques distinctes entre GE et ETI entre 2017 et 2019 donc analyse de plus pour comparer anne par anne """

import matplotlib.pyplot as plt
df_filtered = df[df['annee'] >= 2014]

# Moyennes
avg_ETI_beneficiaire = df_filtered['ETI_beneficiaire_nbr'].mean()
avg_GE_beneficiaire = df_filtered['GE_beneficaire_nbr'].mean()
avg_ETI_dps = df_filtered['ETI_dps_declare'].mean()
avg_GE_dps = df_filtered['GE_dps_declare'].mean()
avg_ETI_CIR = df_filtered['ETI_CIR'].mean()
avg_GE_CIR = df_filtered['GE_CIR'].mean()

# Normalisation des donnes par la moyenne
df_normalized = df_filtered.copy()
df_normalized['ETI_beneficiaire_nbr'] = df_filtered['ETI_beneficiaire_nbr'] - avg_ETI_beneficiaire
df_normalized['GE_beneficaire_nbr'] = df_filtered['GE_beneficaire_nbr'] - avg_GE_beneficiaire
df_normalized['ETI_dps_declare'] = df_filtered['ETI_dps_declare'] - avg_ETI_dps
df_normalized['GE_dps_declare'] = df_filtered['GE_dps_declare'] - avg_GE_dps
df_normalized['ETI_CIR'] = df_filtered['ETI_CIR'] - avg_ETI_CIR
df_normalized['GE_CIR'] = df_filtered['GE_CIR'] - avg_GE_CIR

# Tracer les graphiques
plt.figure(figsize=(18, 6))

# Graphique 1 : Moyenne des bnficiaires d'ETI et GE
plt.subplot(1, 3, 1)
plt.plot(df_filtered['annee'], df_normalized['ETI_beneficiaire_nbr'], marker='o', label='Nombre de bnficiaires ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_beneficaire_nbr'], marker='o', label='Nombre de bnficiaires GE', color='orange')
plt.xlim(2016, 2020)
plt.title("volution normalise des bnficiaires")
plt.xlabel("Anne")
plt.ylabel("Nombre de bnficiaires")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 2 : Moyenne des dpenses dclares d'ETI et GE
plt.subplot(1, 3, 2)
plt.plot(df_filtered['annee'], df_normalized['ETI_dps_declare'], marker='o', label='Dpenses dclars des ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_dps_declare'], marker='o', label='Dpenses dclars des GE', color='orange')
plt.xlim(2016, 2020)
plt.title("volution normalise des dpenses dclares")
plt.xlabel("Anne")
plt.ylabel("Dpenses en M")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

# Graphique 3 : Moyenne des montants de CIR reus
plt.subplot(1, 3, 3)
plt.plot(df_filtered['annee'], df_normalized['ETI_CIR'], marker='o', label='CIR des ETI', color='blue')
plt.plot(df_filtered['annee'], df_normalized['GE_CIR'], marker='o', label='CIR des GE', color='orange')
plt.xlim(2016, 2020)
plt.title("volution normalise du CIR reu")
plt.xlabel("Anne")
plt.ylabel("Crances en M")
plt.xticks(df_filtered['annee'])
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
NAF_dict = {
	"C20_C21": "Pharmacie, parfumerie et entretien",
	"C30": "Construction navale, aronautique et ferroviaire",
	"C29": "Industrie automobile",
	"C26": "Industrie lectrique et lectronique",
	"C22": "Chimie, caoutchouc, plastiques",
	"C25": "Industrie mcanique",
	"C24": "Mtallurgie et transformation des mtaux",
	"C19": "Hydrocarbures, production d'nergie",
	"C13_C14_C15": "Textile, habillement, cuir",
	"C10": "Industrie agricole et alimentaire",
	"C32": "Autres industries manufacturires",
	"C33": "Gestion des holdings industriels",
	"J62": "Conseil et assistance en informatique",
	"M71": "Services d'architecture et d'ingnierie",
	"G": "Commerce",
	"M70": "Conseil et assistance aux entreprises",
	"J61": "Services de tlcommunications",
	"K": "Services bancaires et assurances",
	"M72": "Recherche et dveloppement",
	"S": "Autres services", 
	"A_F": "Autres secteurs" #(Agriculture, sylviculture, pche et Btiment, travaux publics)
}
```

```python
# Charger le fichier CSV
df = pd.read_csv("sector_CIR_evolution.csv", sep=';')

# Supprimer les espaces des noms de colonnes
df.columns = df.columns.str.strip()
# Remplacer les NaN par 0
df.fillna(0, inplace=True)
# Remplacer les virgules par des points dans les colonnes, si ncessaire
df = df.replace(',', '.', regex=True)
# Convertir les colonnes en numriques, en forant les erreurs  NaN
df = df.apply(pd.to_numeric, errors='coerce')

means= {}
for col in df.columns[1:] : 
    means[col] = df[col].mean()

# Initialiser les listes pour les labels et les valeurs
labels = []
values = []

# Extraire les cls et valeurs
for key, value in means.items():
    matched = False  # Indicateur pour vrifier si une correspondance a t trouve
    for naf_key, naf_value in NAF_dict.items():
        if naf_key in key:
            labels.append(naf_value)  # Ajouter le nom correspondant du dict NAF
            values.append(value)  # Ajouter la valeur correspondante
            matched = True
            break
    if not matched:
        print(f"Aucune correspondance trouve pour la cl : {key}")  # Pour le dbogage

# Vrifier les longueurs des listes
print(f"Nombre de labels : {len(labels)}")
print(f"Nombre de valeurs : {len(values)}")


#Affichage
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='skyblue')

# Ajouter un titre et des labels
plt.title('Moyennes des part de CIR  chaque secteur entre 2009 et 2021', fontsize=14)
plt.ylabel('Moyenne des parts de CIR', fontsize=12)

# Faire pivoter les labels de l'axe x pour une meilleure lisibilit
plt.xticks(rotation=45, ha='right')

# Afficher le diagramme
plt.tight_layout()
plt.show()
```

```python
#Filtrer les colonnes avec des moyennes suprieures  8%
filtered_columns = [key for key, value in means.items() if value > 8]

# Afficher l'volution pour les colonnes filtres
plt.figure(figsize=(12, 8))

for col in filtered_columns:
    plt.plot(df['annee'], df[col], marker='o', label=col)
    
# Ajouter des titres et des labels
plt.title('volution des parts de CIR pour les secteurs avec des moyennes > 8%', fontsize=16)
plt.xlabel('Anne', fontsize=12)
plt.ylabel('Part de CIR', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()

# Afficher le graphique
plt.show()
```
